```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
 numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigur,benimEksen)= plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2, color= "#3A95A8") #RGB hex code
benimEksen.plot(numpyDizisi2,numpyDizisi1,"r", alpha=0.2) #alpha saydamlık için
```

```python
(yeniFigur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color="y", linewidth=10.0)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, color="b", linewidth=6.0)

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color="g", linestyle="-.")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color="g", linestyle=":")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 6, color="g", linestyle="--")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7, color="#000000", linestyle="--", marker= "o", markerfacecolor="r")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 8, color="#000000", linestyle="--", marker= "+", markersize= 20)

```

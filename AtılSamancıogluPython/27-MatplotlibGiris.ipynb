```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline 
#jupyter notebook'a özel bir kod. grafikleri göstermek için
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"y")
#en sağa yazdığımız string değeri rengi belirler.
plt.xlabel("Yas Verisi")
plt.ylabel("Kilo Verisi")
plt.title("KİLO'NUN YAŞ'A GÖRE DEĞİŞİM GRAFİĞİ")
plt.show()
#başka bir editörde çalıştırıyorsan show fonksiyonunu kullanacaksın.
```

```python
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
numpyDizisi2
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "g") #KONVEKS
```

```python
plt.plot(numpyDizisi2, numpyDizisi1, "g") #KONKAV
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "g*")
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "g*-")
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "b*--")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1, numpyDizisi2, "r*-")
plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak, şuan 2.grafiği çiziyorum.
plt.plot(numpyDizisi2, numpyDizisi1, "g--")
 
```

```python
benimFigure = plt.figure() 
```

```python
benimFigure = plt.figure() 
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9]) #yükseklik genişlik ayarlamak için 3. ve 4.
figureAxes.plot(numpyDizisi1, numpyDizisi2, "g*")
figureAxes.set_xlabel("X ekseni Veri Ismi")
figureAxes.set_ylabel("Y ekseni Veri Ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.3,0.4,0.2,0.2])

eksen1.plot(numpyDizisi1, numpyDizisi2, "g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2, numpyDizisi1, "b")
eksen2.set_xlabel("X ekseni küçük")
eksen2.set_ylabel("Y ekseni küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(benimFigure,benimEksen)=plt.subplots()
```

```python
(benimFigure, benimEksenler)= plt.subplots(nrows=1,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1, numpyDizisi2, "g")
    eksen.set_xlabel("X EKSENİ")


plt.tight_layout()
    
```

```python
yeniFigur = plt.figure(dpi=100)

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1, numpyDizisi1**2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1**3, label = "numpyDizisi ** 3" )
yeniEksen.legend(loc=1)
```
